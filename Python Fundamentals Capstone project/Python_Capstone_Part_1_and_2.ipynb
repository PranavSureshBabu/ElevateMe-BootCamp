```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**TASK - 1 Data Ingestion Cleaning and Basic Exploration**

```python
class RealEstateDataset:
  def load_data(self, filepath):
    self.filepath = pd.read_csv('property_data.csv')
    return self.filepath.head()

  def clean_data(self):
    df_null = self.filepath.isnull().sum()
    return df_null

  def describe_data(self):
    df_describe = self.filepath.describe()
    return df_describe

  def distribution_property_type(self):
    df_property_type = self.filepath['Type'].value_counts()
    return df_property_type

  def average_price_type(self):
    avg_price_by_type = self.filepath.groupby('Type')['Price'].mean()
    return avg_price_by_type

  def average_size_location(self):
    avg_size_by_location = self.filepath.groupby('Location')['Size_sqft'].mean()
    return avg_size_by_location
```

**TASK - 2 Data Transformation Visualization and Insights**

```python
from datetime import datetime
```

```python
class Analysis:
  def __init__(self, df):
    self.df = df

  def calculate_price_per_sqft(self):
    self.df['Price_per_sqft'] = self.df['Price']/self.df['Size_sqft']
    return self.df.head()

  def calculate_property_age(self):
    current_year = datetime.now().year
    self.df['Property_Age'] = current_year - self.df['Year_Built']
    return self.df.head()
```

```python
#df_property = RealEstateDataset()
data = df_estate.load_data('property_data.csv')
```

```python
df_analysis = Analysis(data)
```

```python
df_analysis.calculate_price_per_sqft()
```

```python
df_analysis.calculate_property_age()
```

```python
class Visualization(Analysis):
  def __init__(self, df: pd.DataFrame):
        self.df = df.copy()

  def plot_price_distribution_by_type(self):
    sns.boxplot(x='Type', y='Price', data=self.df)
    plt.title('Price Distribution by Property Type')
    plt.xlabel('Property Type')
    plt.ylabel('Price')
    plt.show()

  def plot_top_locations(self):
    top_locations = self.df['Location'].value_counts().head(10)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=top_locations.index, y=top_locations.values, palette='Blues_d')
    plt.title('Top 5 Property Locations')
    plt.xlabel('Location')
    plt.ylabel('Count')
    plt.show()

  def show_top_locations(self, n=5):
    top_locations = self.df['Location'].value_counts().head(n)
    print(top_locations)

  def plot_price_trends(self):
    avg_price_by_year = self.df.groupby('Year_Built')['Price'].mean().reset_index()
    plt.figure(figsize=(10, 5))
    sns.lineplot(x='Year_Built', y='Price', data=avg_price_by_year, marker='o')
    plt.title('Average Property Price Trends by Year')
    plt.xlabel('Year')
    plt.ylabel('Average Price')
    plt.grid(True)
    plt.show()

  def plot_price_per_sqft_vs_property_size(self):
    plt.scatter(self.df['Size_sqft'], self.df['Price_per_sqft'], alpha=0.5)
    plt.title('Price per Sqft vs Property Size')
    plt.xlabel('Property Size (sqft)')
    plt.ylabel('Price per Sqft')
    plt.grid(True)
    plt.tight_layout()
    plt.show()



  def plot_mean_price_per_sqft_by_type(self):
    mean_price_per_sqft = self.df.groupby('Type')['Price_per_sqft'].mean().sort_values(ascending=False)
    return mean_price_per_sqft
```

```python
analysis = Analysis(data)
df_with_price = analysis.calculate_price_per_sqft()
df_with_age = analysis.calculate_property_age()
```

```python
visual = Visualization(df_with_price)
```

```python
visual.plot_price_distribution_by_type()
```

```python
visual.show_top_locations()
```

```python
visual.plot_top_locations()
```

```python
visual.plot_price_trends()
```

```python
visual.plot_price_per_sqft_vs_property_size()
```
